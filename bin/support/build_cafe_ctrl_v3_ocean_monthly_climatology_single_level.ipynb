# This notebook saves cafe v3 control daily climatologies in a single dataset

#### IMPORTANT: If adapting this code to save other climatologies, be careful to only include full years, as pyLatte will compute monthly climatologies from the saved daily climatologies

```python
import pandas as pd
import xarray as xr
import numpy as np
from ipywidgets import FloatProgress
from dateutil.relativedelta import relativedelta
```

```python
import sys
sys.maxsize
```

#### Only use last 100 years

```python
base = '/g/data1/v14/coupled_model/v3/OUTPUT/'
years = range(400,500)

paths = []
for year in years:
    path = base + 'ocean_month_0' + str(year) + '_01.nc'
    paths.append(path)
   
ds = xr.open_mfdataset(paths, autoclose=True)
```

```python
ds['sst'] = ds.temp.sel(st_ocean=5.0,drop=True)
```

```python
monthly_av = []
for mn in range(12):
    print(ds.isel(time=range(mn,len(ds.time),12)).time)
    monthly_av.append(ds.isel(time=range(mn,len(ds.time),12)).mean(dim='time'))

ds_clim = xr.concat(monthly_av,dim='time')
for_time = xr.open_dataset('/g/data1/v14/forecast/v1/yr2016/mn1/OUTPUT.1/ocean_month_2016_01.nc')
ds_clim['time'] = for_time.time.astype('<M8[M]').values
ds_clim.time.attrs['long_name'] = 'time'
ds_clim.time.attrs['cartesian_axis'] = 'T'
ds_clim.time.attrs['calendar_type'] = 'JULIAN'
ds_clim.time.attrs['bounds'] = 'time_bounds'
ds_clim = ds_clim.drop('average_DT').drop('time_bounds').drop('temp') \
                 .drop('u').drop('v').drop('wt').drop('salt') \
                 .drop('ty_trans').drop('cfc_11').drop('cfc_12') \
                 .drop('st_edges_ocean').drop('sw_ocean').drop('sw_edges_ocean') \
                 .drop('nv')
```

```python
ds_clim
```

```python
# Save and delete -----
savename = 'cafe.ctrl.v3.ocean_month.sfc.400-499.clim.nc'

ds_clim.to_netcdf(savename,
                  mode = 'w',
                  encoding = {'time':{'dtype':'float','calendar':'JULIAN',
                                      'units':'days since 1870-01-01 00:00:00'}}) 
```
